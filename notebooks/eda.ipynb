#  Unemployment Analysis in Tunisia

## Objective
This project analyzes unemployment trends in Tunisia using World Bank data.

The goal is to:
- Understand how unemployment evolved over time
- Compare male and female unemployment
- Analyze youth unemployment trends
- Identify key patterns and structural issues in the labor market

```python
import pandas as pd
total = pd.read_csv("../data/total.csv", skiprows=4)
male = pd.read_csv("../data/male.csv", skiprows=4)
female = pd.read_csv("../data/female.csv", skiprows=4)
youth = pd.read_csv("../data/youth.csv", skiprows=4)

total = total[total["Country Code"] == "TUN"]
male = male[male["Country Code"] == "TUN"]
female = female[female["Country Code"] == "TUN"]
youth = youth[youth["Country Code"] == "TUN"]
```

```python
def reshape(df, value_name):
    df = df.melt(
        id_vars=["Country Name", "Country Code", "Indicator Name"],
        var_name="year",
        value_name=value_name
    )
    
    df = df[df["year"].str.isnumeric()]
    df["year"] = df["year"].astype(int)
    
    return df[["year", value_name]]
total = reshape(total, "total_unemployment")
male = reshape(male, "male_unemployment")
female = reshape(female, "female_unemployment")
youth = reshape(youth, "youth_unemployment")
```

##  Data Cleaning

Steps performed:
- Removed missing values
- Sorted data by year
- Reset index

```python
df = total.merge(male, on="year", how="inner")
df = df.merge(female, on="year", how="inner")
df = df.merge(youth, on="year", how="inner")
df = df.dropna()
df = df.sort_values("year")

df = df.reset_index(drop=True)

print(df.head())
print(df.info())
```

```python
df.to_csv("../data/tunisia_unemployment_clean.csv", index=False)
df.head()
df.info()
```

###  Unemployment Trend Over Time

This chart shows how unemployment has evolved in Tunisia.

```python
import plotly.express as px

fig = px.line(
    df,
    x="year",
    y="total_unemployment",
    title="Unemployment Trend in Tunisia"
)

fig.show()
```

**Insight:**

The unemployment rate shows fluctuations over time with an overall increasing trend, suggesting economic instability and sensitivity to external factors.

### 👥 Gender Comparison

This analysis compares unemployment between males and females.

```python
fig = px.line(
    df,
    x="year",
    y=["male_unemployment", "female_unemployment"],
    title="Gender Gap in Unemployment (Tunisia)"
)

fig.show()
```

**Insight:**

Female unemployment tends to be higher than male unemployment, indicating potential inequalities in access to employment opportunities.

###  Youth Unemployment

This chart focuses on unemployment among young people.

```python
fig = px.line(
    df,
    x="year",
    y="youth_unemployment",
    title="Youth Unemployment in Tunisia"
)

fig.show()
```

### ⚖️ Youth vs Total Unemployment

This comparison highlights the gap between youth and overall unemployment.

```python
fig = px.line(
    df,
    x="year",
    y=["total_unemployment", "youth_unemployment"],
    title="Youth vs Total Unemployment"
)

fig.show()

```

**Insight:**

The persistent gap between youth and total unemployment suggests difficulties in transitioning from education to employment.

###  Gender Gap Analysis

```python
df["gender_gap"] = df["female_unemployment"] - df["male_unemployment"]

fig = px.line(df, x="year", y="gender_gap", title="Gender Gap (Female - Male)")
fig.show()
```

**Insight:**

A positive gender gap indicates that female unemployment is higher than male unemployment, reinforcing the presence of gender disparities.

##  Conclusion

Key findings from this analysis:

- Unemployment in Tunisia shows an overall increasing trend over time.
- Female unemployment tends to be higher than male unemployment.
- Youth unemployment is significantly higher than the national average.
- The labor market faces structural challenges, particularly for young people.

This analysis highlights the need for targeted policies to improve employment opportunities, especially for youth and women.
